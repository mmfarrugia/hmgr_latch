```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn
import pandas as pd

seaborn.set_theme(rc={'figure.figsize':(11.7,8.27)})
```

```python
results = pd.read_csv('output_saturation-results.csv')
results = results[results['scored_state']=='ddG']
results = results[results['score_function_name'] == 'fa_talaris2014-gam']
results
```

```python
fig, ax = plt.subplots(1,1,figsize=(20,8))
seaborn.boxplot(data=results[results['scored_state']=='ddG'], x='case_name', y='total_score', ax=ax)
ax.tick_params(axis='x', rotation=30)
```

```python
last_results = results[results['nstruct'] >= 30]
last_results
```

```python
fig, ax = plt.subplots(1,1,figsize=(20,8))
seaborn.boxplot(data=last_results[last_results['scored_state']=='ddG'], x='case_name', y='total_score', ax=ax)
ax.tick_params(axis='x', rotation=30)
```

```python
fig, ax = plt.subplots(1,1,figsize=(20,8))
seaborn.swarmplot(data=results[results['scored_state']=='ddG'], x='case_name', y='total_score', ax=ax)
ax.tick_params(axis='x', rotation=30)
```

```python
fig, ax = plt.subplots(1,1,figsize=(20,8))
seaborn.swarmplot(data=last_results[last_results['scored_state']=='ddG'], x='case_name', y='total_score', ax=ax)
ax.tick_params(axis='x', rotation=30)
```

```python
fig, ax = plt.subplots(1,1,figsize=(20,8))
seaborn.barplot(data=last_results, x='case_name', y='total_score', ci="sd", ax=ax)
ax.tick_params(axis='x', rotation=90)
```

```python
fig, ax = plt.subplots(1,1,figsize=(20,8))
average_by_mutation = last_results.groupby('case_name').mean()
seaborn.scatterplot(data=average_by_mutation, x='case_name', y='total_score', ax=ax)
ax.tick_params(axis='x', rotation=90)
```
